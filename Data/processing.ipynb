# Data preparation

```python
import pandas as pd
import calendar
```

```python
df = pd.read_csv('sarawak_visitor_arrival.csv')
print(df.shape)

df.head()
```

```python
df.Year = df.Year.astype(int)
d = {'January':1,
     'February':2,
     'March':3,
     'April':4,
     'May':5,
     'June':6,
     'July':7,
     'August':8,
     'September':9,
     'October':10,
     'November':11,
     'December':12}
df.Month = df.Month.str.strip().map(d)
df.Month = df.Month.astype(int)

df.head()
```

```python
final = pd.melt(df, id_vars=['Year','Month'], value_vars=df.columns, var_name='country', value_name='cnt')
final.head()
```

```python
final.to_csv('data.csv')
```
